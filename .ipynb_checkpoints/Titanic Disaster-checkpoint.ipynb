```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model, model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
print('Modules Imported')
```

```python
test_df = pd.read_csv('Input/test.csv')
train_df = pd.read_csv('Input/train.csv',index_col=0)
```

```python
train_df.head()
```

```python
train_df.describe()
```

```python
train_df.dropna(subset=['Age','Embarked'],inplace=True)
train_df.info()
```

```python
def map_Embarked(e):
    if e == 'C':
        return 0
    elif e == 'Q':
        return 1
    elif e == 'S':
        return 2

def map_Sex(sex):
    if sex == 'male':
        return 0
    elif sex == 'female':
        return 1
    
train_df['Embarked'] = train_df.Embarked.apply(lambda e : map_Embarked(e))
train_df['Sex'] = train_df.Sex.apply(lambda sex : map_Sex(sex))
```

```python
fig , ax = plt.subplots(3,2,figsize=(15,18))

train_df[train_df['Survived']==1].Pclass.value_counts().plot.bar(
    ax = ax[0][0]
)
xlabels = ['1st Class','2nd Class','3rd Class']
ax[0][0].set_title('Passengers Survived classified according to Class')
ax[0][0].set_xticklabels(xlabels,rotation='horizontal')
ax[0][0].set_xlabel('Passenger Class')
ax[0][0].set_ylabel('Number of Passengers')

train_df[train_df['Survived']==1].Sex.value_counts().plot.bar(
    ax = ax[0][1]
)
ax[0][1].set_title('Passengers Survived classified according to Sex')
ax[0][1].set_xticklabels(['Female','Male'],rotation='horizontal')
ax[0][1].set_xlabel('Sex')
ax[0][1].set_ylabel('Number of Passengers')

Age_Series = train_df[train_df['Survived']==1].Age.value_counts()
Age_Series.sort_index(inplace=True)
Age_Series.plot(kind='line',ax=ax[1][0])
ax[1][0].set_title('Passengers Survived classified according to Age')
ax[1][0].set_xlabel('Age')
ax[1][0].set_ylabel('Number of Passengers')

SibSp_Series = train_df[train_df['Survived']==1].SibSp.value_counts()
SibSp_Series.sort_index(inplace=True)
SibSp_Series.plot(kind='line',ax=ax[1][1],label='Siblings')

Parch_Series = train_df[train_df['Survived']==1].Parch.value_counts()
Parch_Series.sort_index(inplace=True)
Parch_Series.plot(kind='line',ax=ax[1][1],label='Parch')

ax[1][1].set_title('Passengers Survived classified according to # of Siblings and Parch Aboard')
ax[1][1].set_xlabel('Number of People')
ax[1][1].set_ylabel('Number of Passengers')
ax[1][1].legend()

train_df[train_df['Survived']==1].Fare.value_counts().sort_index().plot(kind='line',ax=ax[2][0])
ax[2][0].set_title('Passengers Survived classified according to Fare')
ax[2][0].set_xlabel('Fare')
ax[2][0].set_ylabel('Number of Passengers')

train_df[train_df['Survived']==1].Embarked.value_counts().plot(kind='bar',ax=ax[2][1])
ax[2][1].set_title('Passengers Survived classified according to boarding point')
ax[2][1].set_xlabel('Place')
ax[2][1].set_ylabel('Number of Passengers')
xlabels = ['Southampton','Cherbourg','Queenstown']
ax[2][1].set_xticklabels(xlabels,rotation='horizontal');
```

```python
X = train_df.loc[:,['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
y = train_df.loc[:,['Survived']]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
values_C = [0.01,0.03,0.1,0.3,1,3,10,100]

for i in values_C:
    clf = LogisticRegression(C=i,random_state=1,solver='liblinear')
    clf.fit(X_train,np.ravel(y_train))
    result = clf.predict(X_test)
    accuracy = (result == y_test['Survived']).sum()/(result == y_test['Survived']).count()
    print('C = ' + str(i) + ' Accuracy = ' + str(accuracy))
```

```python
clf = LogisticRegression(C=0.1,random_state=1,solver='liblinear')
clf.fit(X_train,np.ravel(y_train));
```

```python
test_df['Age'].fillna(np.mean(test_df['Age']),inplace=True)
test_df['Fare'].fillna(np.mean(test_df['Fare']),inplace=True)
test_df['Sex'] = test_df.Sex.apply(lambda sex : map_Sex(sex))
test_df['Embarked'] = test_df.Embarked.apply(lambda e : map_Embarked(e))
```
